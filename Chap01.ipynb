1-1. AL / ML / DL의 차이

인공지능 : 사람처럼 학습하고 추론할 수 있는 지능을 가진 컴퓨터 시스템을 만드는 기술
    인공일반지능/강인공지능 : 사람과 구분하기 어려운 지능을 가진 컴퓨터 시스템
    약인공지능 : 특정 분야에서 사람의 일을 도와주는 보조 역할

머신러닝 : 규칙을 일일이 프로그래밍하지 않아도 자도응로 데이터에서 규칙을 학습하는 알고리즘을 연구하는 분야
        : 인공지능의 하위 분야 중 지능을 구현하기 위한 소프트웨어어를 담당하는 핵심분야
    사이킷런 : 컴퓨터 과학 분야의 대표적인 머신러닝 라이브러리 ( 파이썬 API 기반 )

딥러닝 : 인공신경망을 기반으로 한 방법 통칭
    Tensorflow , Pytorch : 오픈소스 딥러닝 라이브러리 

1-2. 코랩 / 쥬피터 노트북

Colab : 웹 브라우저에서 무료로 파이썬 프로그램을 테스트하고 저장할 수 있는 서비스
      : 클라우드 기반의 주피터 노트북 개발환경     
Colab notebook : 웹 브라우저에서 텍스트와 프로그램 코드를 자유롭게 작성할 수 있는 온라인 에디터
               : 구글 클라우드의 가상서버(VM)를 사용
               : 구글 클라우드의 컴퓨트 엔진에 연결 ( MEMORY 12GB / DISK SPACE 225GB 사용가능 ) 
    cell : 코드 또는 텍스트의 덩어리 
         : 실행할 수 있는 최소 단위
         : code cell / text cell
